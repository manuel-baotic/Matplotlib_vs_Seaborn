### Libraries importeren

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

### Daten laden

```python
exercise = sns.load_dataset("exercise")
exercise
```

### Daten aussortieren mittels Logikvektor

```python
lv = (exercise['kind'] == 'running')
selected_exercise = exercise[lv]
selected_exercise
```

### Histogramm in Seaborn kind = running

```python
sns.set()
plt.xlabel("pulse")
plt.ylabel("Observations")
plt.hist(selected_exercise['pulse'], normed=False, alpha=0.5);
```

### Daten selektieren kind = 'rest'

```python
lv = (exercise['kind'] == 'rest')
selected_exercise = exercise[lv]
selected_exercise
```

### Histogramm in Matplotlib kind = rest

```python
sns.reset_orig() 
ax = plt.gca()
ax.grid(True)
plt.xlabel("pulse")
plt.ylabel("Observations")
plt.hist(selected_exercise['pulse'], normed=False, alpha=0.5);
```

### 'tips' Daten laden 

```python
tips = sns.load_dataset("tips")
tips
```

```python
#Seaborn setzen
sns.set()

#Schriftögrße setzen dazu wird immer ein Kontext benötigt
sns.set_context('paper',font_scale=1.5)

#hue ist die Einfärbung, size die Plotgröße
g = sns.FacetGrid(tips,col="day", hue="time", size=3)

#Parameter alpha für Transparenz
g.map(plt.scatter, "total_bill", "tip", alpha=.5)

#fügt Legende hinzu
g.add_legend();
```
